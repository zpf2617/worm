```python
#发送邮件类
import smtplib
from email.message import EmailMessage
from datetime import datetime

class emailclass:
    def __init__(self,content="Nothing") -> None:
        self.address="ztoz@foxmail.com"
        self.target="775437720@qq.com"
        self.authcode="casqvbynnexobfib"
        self.content=content
        self.Subject="python emailclass send_email"
        self.host="smtp.qq.com"
        self.port=465
        self.msg=EmailMessage()

    def set_email(self)->str:
        self.msg["From"] = self.address
        self.msg["To"] = self.target
        self.msg["Subject"] = self.Subject
        self.msg.set_content(self.content)
        return str(self.msg)

    def send_email(self):
        self.set_email()
        try:
            with smtplib.SMTP_SSL(self.host,self.port) as smtp:
                smtp.login(self.address,self.authcode)#邮箱，授权码
                smtp.send_message(self.msg)
                print("发送正常。")
        except Exception as e:#捕捉全部异常
            print("发生异常，异常信息：",repr(e)) #翻译异常信息


if __name__ == "__main__":
    emailclass(content=str(datetime.now())).send_email()
    print(emailclass().set_email())
```
