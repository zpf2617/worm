```python
from tkinter import *
import tkinter.filedialog

from docx import Document
import xlwt
import re



root=tkinter.Tk()
root.title("试卷题库提取器")
# 设置窗口位置
root.geometry("+500+300")
# 设置窗口大小
root.geometry('400x400')
# 设置图标，与py文件放在一个文件夹下
root.iconbitmap('b747421112.ico')




# 下面这个tkinter.StringVar()比较关键，不用他无法输出正确的文件路径
filename = tkinter.StringVar()
def opent():
    filepath=tkinter.filedialog.askopenfilename()
    filename.set(filepath)
    lb.config(text=filename)

def doexcel():
    lb.config(text='OK!')
    xl = xlwt.Workbook(encoding='utf-8')
    # 创建一个sheet1对象,第二个参数是指单元格是否允许重设置，默认为False
    sheet1 = xl.add_sheet('单选题', cell_overwrite_ok=True)
    sheet2 = xl.add_sheet('多选题', cell_overwrite_ok=True)
    sheet3 = xl.add_sheet('判断题', cell_overwrite_ok=True)
    sheet4 = xl.add_sheet('问答题', cell_overwrite_ok=True)
    # 初始化样式
    style = xlwt.XFStyle()
    # 为样式创建字体
    font = xlwt.Font()
    font.name = 'Times New Roman'
    # 黑体
    font.bold = True
    # 下划线
    font.underline = False
    # 斜体字
    font.italic = True
    # 设定样式
    style.font = font
    # 第一个参数代表行，第二个参数是列，第三个参数是内容，第四个参数是格式
    # 初始化
    sheet1.write(0, 0, '序号', style)
    sheet1.write(0, 1, '题目', style)
    sheet1.write(0, 2, 'A', style)
    sheet1.write(0, 3, 'B', style)
    sheet1.write(0, 4, 'C', style)
    sheet1.write(0, 5, 'D', style)

    sheet2.write(0, 0, '序号', style)
    sheet2.write(0, 1, '题目', style)
    sheet2.write(0, 2, 'A', style)
    sheet2.write(0, 3, 'B', style)
    sheet2.write(0, 4, 'C', style)
    sheet2.write(0, 5, 'D', style)

    sheet3.write(0, 0, '序号', style)
    sheet3.write(0, 1, '题目', style)

    sheet4.write(0, 0, '序号', style)
    sheet4.write(0, 1, '题目', style)
    # sheet1.write(1, 0, '带样式的写入', style)


    # 段落计数器
    paragraph=0

    # 选项计数器
    # abcd=0
    # 行
    item=0
    title=0
    a,b,c,d,e,f=0,0,0,0,0,0

    # 读取DOCX试卷文本
    # 这里出现了问题！  filenames返回路径式/斜杠，而Document(是\反斜杠地址)
    # 终于成功了，filename = tkinter.StringVar()  这个用法注意
    doc1=Document(str(filename.get()))
    # doc1=Document(r'C:\Users\zero\.ipython\试卷题库提取器\test.docx')
    for p in doc1.paragraphs:
        
        paragraph+=1
    

        print(p.text)
    
        print(title)
        if re.search(r'单选题', p.text):       #.span()
            title=1
            item=0
        elif re.search(r'多选题', p.text):       #.span()
            title=2
            item=0
        elif re.search(r'判断题', p.text):       #.span()
            title=3
            item=0
        elif re.search(r'问答题', p.text):       #.span()
            title=4
            item=0
        elif re.search(r'填空题', p.text):       #.span()
            title=5
            item=0
        else :
            pass

        # result1 = re.findall(r'\d{1,2}\.\w+\s',p.text)
        # 现在开始处理单选题1-100
        if title == 1:
            # 查找数字开头的题目并返回查找结果
            result1 = re.findall(r'^\d{1,3}\..*',p.text)
            
            if result1:
                item=item+1
                # 如果找到单选题条目的题目内容
                for i in result1:
                    # 写入序号
                    sheet1.write(item, 0, item)
                    #替换题目开头的数字
                    sub1 = re.sub(r'^\d{1,3}\.', "+",i)
                    sheet1.write(item, 1, sub1)
                    # abcd=0
                
            else :
                # 用了4个分隔区分AB选项
                result_A = re.findall(r'A\..*    ',p.text)
                result_B = re.findall(r'B\..*',p.text)
                result_C = re.findall(r'C\..*    ',p.text)
                result_D = re.findall(r'D\..*',p.text)
                
                for a in result_A:
                    sub1 = re.sub(r'A\.', "",a)
                    sheet1.write(item, 2, sub1)
                
            
                for b in result_B:
                    sub1 = re.sub(r'B\.', "",b)
                    sheet1.write(item, 3, sub1)

            
                for c in result_C:
                    sub1 = re.sub(r'C\.', "",c)
                    sheet1.write(item, 4, sub1)

            
                for d in result_D:
                    sub1 = re.sub(r'D\.', "",d)
                    sheet1.write(item, 5, sub1)
        

        # ------------------------------------------------------------------------------------
        # 现在开始处理多选题
        elif title == 2:
            # 查找数字开头的题目并返回查找结果
            result1 = re.findall(r'^\d{1,3}\..*',p.text)
            
            if result1:
                item=item+1
                # 如果找到单选题条目的题目内容
                for i in result1:
                    # 写入序号
                    sheet2.write(item, 0, item)
                    #替换题目开头的数字
                    sub1 = re.sub(r'^\d{1,3}\.', "+",i)
                    sheet2.write(item, 1, sub1)
                    # abcd=0
                
            else :
                # 用了4个分隔区分AB选项
                result_A = re.findall(r'A\..*    ',p.text)
                result_B = re.findall(r'B\..*',p.text)
                result_C = re.findall(r'C\..*    ',p.text)
                result_D = re.findall(r'D\..*',p.text)
                
                for a in result_A:
                    sub1 = re.sub(r'A\.', "",a)
                    sheet2.write(item, 2, sub1)
                
            
                for b in result_B:
                    sub1 = re.sub(r'B\.', "",b)
                    sheet2.write(item, 3, sub1)

            
                for c in result_C:
                    sub1 = re.sub(r'C\.', "",c)
                    sheet2.write(item, 4, sub1)

            
                for d in result_D:
                    sub1 = re.sub(r'D\.', "",d)
                    sheet2.write(item, 5, sub1)         
                
        # ------------------------------------------------------------------------------------
        # 现在开始处理判断题
        elif title == 3:
            # 查找数字开头的题目并返回查找结果
            result1 = re.findall(r'^\d{1,3}\..*',p.text)
            
            if result1:
                item=item+1
                # 如果找到单选题条目的题目内容
                for i in result1:
                    # 写入序号
                    sheet3.write(item, 0, item)
                    #替换题目开头的数字
                    sub1 = re.sub(r'^\d{1,3}\.', "+",i)
                    sheet3.write(item, 1, sub1)

        # ------------------------------------------------------------------------------------
        # 现在开始处理问答题
        elif title == 4:
            # 查找数字开头的题目并返回查找结果
            result1 = re.findall(r'^\d{1,3}\..*',p.text)
            
            if result1:
                item=item+1
                # 如果找到单选题条目的题目内容
                for i in result1:
                    # 写入序号
                    sheet4.write(item, 0, item)
                    #替换题目开头的数字
                    sub1 = re.sub(r'^\d{1,3}\.', "+",i)
                    sheet4.write(item, 1, sub1)

    # 保存文件,保存在当前目录下
    xl.save('Re_test试卷题库提取器结果.xls')




lb=Label(root,text="filename")
lb.pack()

btn=Button(root,text="请选择试卷！",command=opent)
btn.pack()

btn=Button(root,text="提取.docx试卷到.xls文件",command=doexcel)
btn.pack()


root.mainloop()
```
