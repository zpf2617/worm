```python
#爬虫2：成功，User-Agent伪装
#简单带参数爬虫2网站有反爬虫机制UA:User-Agent伪装，不能趴到数据
import requests

url="https://www.sogou.com/web"

# 关键词字典打包成参数
kw=input("请输入一个需要搜索的内容")
param={
    "query":kw
}

headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36"
}


response=requests.get(url=url,params=param,headers=headers)
print(response.text)


page_text=response.text

fileName=kw+'.html'

with open("D:/"+fileName,'w',encoding='utf-8') as fp:
    fp.write(page_text)

print(fileName,"保存成功")
```

```python
#爬虫4：百度翻译
#简单带参数爬虫3网站有反爬虫机制UA:User-Agent伪装，不能趴到数据
import requests
import json
url="https://fanyi.baidu.com/sug"

# 关键词字典打包成参数
kw=input("请输入需要翻译的英文")
data={
    "kw":kw
}

headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36"
}


response=requests.post(url=url,data=data,headers=headers)
print(response.text)


dic_obj=response.json()

fileName="post-"+kw+'.html'



with open("D:/"+fileName,'w',encoding='utf-8') as fp:
    json.dump(dic_obj,fp=fp,ensure_ascii=False)

print(json.dumps(dic_obj,ensure_ascii=False))
print(fileName,"保存成功")

print(dic_obj['data'][0]["k"])
print(":")
print(dic_obj['data'][0]["v"])
```
